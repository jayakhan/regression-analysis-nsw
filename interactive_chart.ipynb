```python
import pandas as pd

lalonde = pd.read_csv(
    "lalondedata.txt", sep =",")
lalonde.sample(5)
```

```python
lalonde = lalonde.rename(columns={'Unnamed: 0': 'workers'})
list_workers = []
for i in lalonde['workers']:
    if 'NSW' in i:
        list_workers.append(i.replace('NSW', ''))
    else:
        list_workers.append(i.replace('PSID', ''))
lalonde['id'] = list_workers
lalonde.sample(5)
```

```python
lalonde['changeinincome'] = lalonde['re78'] - lalonde['re74']
lalonde.sample(5)
```

```python
list_ms = []
for i in lalonde['married']:
    if i == 1:
        list_ms.append('Married')
    else:
        list_ms.append('Unmarried')
lalonde['ms'] = list_ms
lalonde.sample(5)
```

```python
import altair as alt

ms = ['Married', 'Unmarried']

base = alt.Chart(lalonde, width=400, height=300).mark_point(filled=True).encode(
    x=alt.X('id', scale=alt.Scale(clamp=False), axis=None),
    y='changeinincome:Q',
    tooltip="ms:N"
)
#color changing marks
ms_radio = alt.binding_radio(options=ms)

rating_select = alt.selection_single(fields=['ms'], bind=ms_radio, name="Select")
rating_color_condition = alt.condition(rating_select,
                      alt.Color('ms:N'),
                      alt.value('lightgrey'))

highlight_ratings = base.add_selection(
    rating_select
).encode(
    color=rating_color_condition
).properties(title="Change in income for Married and Unmarried workers")

highlight_ratings
```

```python
list_race = []
for i in lalonde['black']:
    if i == 1:
        list_race.append('Black')
    else:
        list_race.append('Not Black')
lalonde['race'] = list_race
lalonde.sample(5)
```

```python
import altair as alt

race = ['Black', 'Not Black']

base1 = alt.Chart(lalonde, width=400, height=300).mark_point(filled=True).encode(
    x=alt.X('id', scale=alt.Scale(clamp=False), axis=None),
    y='changeinincome:Q',
    tooltip="race:N"
)
#color changing marks
race_radio = alt.binding_radio(options=race)

race_select = alt.selection_single(fields=['race'], bind=race_radio, name="Select ")
race_color_condition = alt.condition(race_select,
                      alt.Color('race:N'),
                      alt.value('lightgrey'))

highlight_race = base1.add_selection(
    race_select
).encode(
    color=race_color_condition
).properties(title="Change in income for Married and Unmarried workers")

highlight_race
```
